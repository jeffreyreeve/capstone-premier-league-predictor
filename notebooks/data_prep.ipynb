```python
import os
import re
import time 
path = '../../Fantasy-Premier-League/data/2022-23/'
understats_columns = ['xG','xGA','ppda','date']
team_understats = {}
files = sorted(os.listdir(path+'understat/'))[1:]
for idx,file in enumerate(files):
    row = pd.read_csv(path+'understat/'+file)[understats_columns]
    for loc in range(row.shape[0]):
        nums = re.findall(r'\d+', row['ppda'].loc[loc])
        res = int(nums[0]) / int(nums[1])
        row['ppda'].loc[loc] = res
    team_understats[teams[idx]] = row
team_understats
```
