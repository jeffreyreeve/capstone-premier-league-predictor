```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
#from ELO import ELO_rating
```

```python
# Read in the PL22-23.csv file from https://www.football-data.co.uk/englandm.php
pd.set_option('display.max_columns', None)
season23 = pd.read_csv('../data/PL22-23.csv')
season23.head()
```

```python
season23 = season23.drop(['Div', 'FTR'], axis=1)
season23.head()
```

```python
season23.rename(columns={'HomeTeam': 'TeamH', 'AwayTeam': 'TeamA'}, inplace=True)
season23.head()
```

```python
season23
```

```python
# Create a new DataFrame to store the transformed data
new_df = pd.DataFrame(columns=['Date', 'Time', 'TeamH', 'FTHG', 'HTHG', 'HS', 'HST', 'HF', 'HC', 'HY', 'HR'])

# Iterate over the rows of the original DataFrame
for _, row in season23.iterrows():
    date = row['Date']
    time = row['Time']
    home_team = row['TeamH']
    away_team = row['TeamA']
    fthg = row['FTHG']
    hthg = row['HTHG']
    hs = row['HS']
    hst = row['HST']
    hf = row['HF']
    hc = row['HC']
    hy = row['HY']
    hr = row['HR']

    # Append a new row for the home team
    new_df = new_df.append({'Date': date, 'Time': time, 'TeamH': home_team, 'FTHG': fthg, 'HTHG': hthg, 'HS': hs, 'HST': hst,
                            'HF': hf, 'HC': hc, 'HY': hy, 'HR': hr}, ignore_index=True)

    # Append a new row for the away team
    new_df = new_df.append({'Date': date, 'Time': time, 'TeamH': away_team, 'FTHG': row['FTAG'], 'HTHG': row['HTAG'],
                            'HS': row['AS'], 'HST': row['AST'], 'HF': row['AF'], 'HC': row['AC'],
                            'HY': row['AY'], 'HR': row['AR']}, ignore_index=True)

# Print the transformed DataFrame
new_df
```

```python
new_df.rename(columns={'TeamH': 'Team', 'FTHG': 'FTG', 'HTHG': 'HTG', 'HS': 'S', 'HST': 'ST', 'HF': 'F', 'HC': 'C', 'HY': 'Y', 'HR': 'R'}, inplace=True)
new_df.head()
```

```python
new_df
```

```python
season23_cleaning = season23.drop(['TeamH', 'TeamA', 'FTHG', 'FTAG', 'HTHG', 'HTAG', 'HTR', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR'], axis=1)
season23_cleaning
```

```python
season23_cleaning = season23_cleaning.loc[season23_cleaning.index.repeat(2)].sort_index().reset_index(drop=True)
season23_cleaning
```

```python
season23_merged = pd.merge(new_df, season23_cleaning, on=['Date', 'Time'], how='left')
season23_merged
```

```python
season23_merged = season23_merged.drop_duplicates()
season23_merged
```
