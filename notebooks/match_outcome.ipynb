```python
def previous_games(Team, GameID):
    
    team_stats = combined_stats[(combined_stats['value'] == Team) & (combined_stats['GameID'] <= GameID)].sort_values(['Date']).tail()[column_names[2:]] 
    GameIDs = combined_stats[(combined_stats['value'] == Team) & (combined_stats['GameID'] <= GameID)].sort_values(['Date']).tail()['GameID']
    opponent_stats = combined_stats[(combined_stats['GameID'].isin(GameIDs)) & (combined_stats['value'] != Team)].sort_values('Date')[column_names[2:]] 
    betting_data = final_data_corrected[(final_data_corrected['GameID'].isin(GameIDs))].sort_values('Date')[betting_stats]
    combined = pd.concat([team_stats, opponent_stats, betting_data], axis=1) 
    
    return combined
```
